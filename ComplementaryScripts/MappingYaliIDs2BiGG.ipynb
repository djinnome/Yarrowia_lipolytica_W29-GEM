# Based on MetaNetX IDs now map to KEGG, BiGG and Biocyc

## Metabolites

```python
from pandas import read_csv
import pandas as pd
from numpy import isnan
```

```python
xref = read_csv("chem_xref.tsv", sep='\t', header=385)
```

```python
xref = xref.loc[xref["#XREF"] != xref["MNX_ID"], :]
# Split namespaces from identifiers.
xref[["XREF_ID", "XREF"]] = xref["#XREF"].str.split(":", n=1, expand=True)
# Group the data in the xref dataframe so that one MNX ID maps to all
# corresponding cross-references from other databases. Then list all
# identifiers that belong to these databases:
# MNX_ID    XREF_ID
# MNXM0     chebi       [23367, 59999]
#           metacyc     [UNKNOWN]
# Make a separate column for every XREF_ID:
# MNX_ID    chebi           metacyc
# MNXM0     [23367, 59999]  [UNKNOWN]
xref = xref.groupby(["MNX_ID", "XREF_ID"], as_index=False, sort=False)[
    "XREF"].apply(list).unstack('XREF_ID')
# Re-insert MetaNetX identifiers as lists.
# FIXME: Shouldn't we use metanetx.chemical here instead of 'mnx'?
xref["mnx"] = [[x] for x in xref.index]
```

```python
chem_mapping = xref.to_dict('index')
```

```python
chem_mapping
```

```python
bigg_count = 0
for met in yali.metabolites:
    if "metanetx.chemical" in met.annotation and \
    met.annotation["metanetx.chemical"] in chem_mapping:
        available_annotation = chem_mapping[met.annotation["metanetx.chemical"]]
        if type(available_annotation["bigg"]) != float:
            met.annotation["bigg.metabolite"] = available_annotation["bigg"][0]
            bigg_count+=1
        if type(available_annotation["kegg"]) != float:
            met.annotation.setdefault("kegg.compound", available_annotation["kegg"])
        if type(available_annotation["metacyc"]) != float:
            met.annotation.setdefault("biocyc", available_annotation["metacyc"])
```

```python
unmapped_mets = []
for met in yali.metabolites:
    if "metanetx.chemical" not in met.annotation:
        unmapped_mets.append(met)
        
print("Metabolites that have no metanet.chemical annotation {}".format(len(unmapped_mets)))
print("Metabolites that have bigg.metabolite annotation {}".format(bigg_count))
print("Total metabolites in Yali {}".format(len(yali.metabolites)))
```

## Reactions

```python
xref_reac = read_csv("reac_xref.tsv", sep='\t', skiprows=386, header=None, names=["#XREF","MNX_ID","Description"])
```

```python
xref_reac = xref_reac.loc[xref_reac["#XREF"] != xref_reac["MNX_ID"], :]

xref_reac[["XREF_ID", "XREF"]] = xref_reac["#XREF"].str.split(":", n=1, expand=True)
# Group the data in the xref dataframe so that one MNX ID maps to all
# corresponding cross-references from other databases. Then list all
# identifiers that belong to these databases:
# MNX_ID    XREF_ID
# MNXM0     chebi       [23367, 59999]
#           metacyc     [UNKNOWN]
# Make a separate column for every XREF_ID:
# MNX_ID    chebi           metacyc
# MNXM0     [23367, 59999]  [UNKNOWN]
xref_reac = xref_reac.groupby(["MNX_ID", "XREF_ID"], as_index=False, sort=False)[
    "XREF"].apply(list).unstack('XREF_ID')
# Re-insert MetaNetX identifiers as lists.
# FIXME: Shouldn't we use metanetx.chemical here instead of 'mnx'?
xref_reac["mnx"] = [[x] for x in xref_reac.index]
```

```python
reac_mapping = xref_reac.to_dict('index')
```

```python
reac_mapping
```

```python
bigg_count_rxn = 0
for rxn in yali.reactions:
    if "metanetx.reaction" in rxn.annotation and \
    rxn.annotation["metanetx.reaction"] in reac_mapping:
        available_annotation = reac_mapping[rxn.annotation["metanetx.reaction"]]
        if type(available_annotation["bigg"]) != float:
            rxn.annotation["bigg.reaction"] = available_annotation["bigg"][0]
            bigg_count_rxn +=1
        if type(available_annotation["kegg"]) != float:
            rxn.annotation.setdefault("kegg.reaction", available_annotation["kegg"])
        if type(available_annotation["metacyc"]) != float:
            rxn.annotation.setdefault("biocyc", available_annotation["metacyc"])
        if type(available_annotation["rhea"]) != float:
            rxn.annotation.setdefault("rhea", available_annotation["rhea"])
        if type(available_annotation["reactome"]) != float:
            rxn.annotation.setdefault("reactome", available_annotation["reactome"])
```

```python
unmapped_rxns = []
for rxn in yali.reactions:
    if "metanetx.reaction" not in rxn.annotation:
        unmapped_rxns.append(rxn)
        
print("Reactions that have no metanet.reaction annotation {}".format(len(unmapped_rxns)))
print("Reactions that have bigg.reaction annotation {}".format(bigg_count_rxn))
print("Total metabolites in Yali {}".format(len(yali.reactions)))
```
